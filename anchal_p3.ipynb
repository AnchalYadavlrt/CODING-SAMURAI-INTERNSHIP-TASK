```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Coding_Samurai\Titanic_train.csv")
df.head()
```

```python
print(df.shape)
```

```python
print(df.columns.tolist())
```

```python
print(df.describe())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.hist(df["Age"], bins=20, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.hist(df["Fare"], bins=20, edgecolor="black")
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Age"])
plt.title("Age Boxplot")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Fare"])
plt.title("Fare Boxplot")
plt.show()
```

```python
numeric_data = df[[
    "PassengerId",
    "Survived",
    "Pclass",
    "Age",
    "SibSp",
    "Parch",
    "Fare"
]]
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("\n----------- Observations -----------")
print("1. Missing values were found only in the Age column.")
print("2. Missing Age values were filled using the median.")
print("3. Female passengers had a higher survival rate than males.")
print("4. First-class passengers survived more frequently.")
print("5. Most passengers were between 20 and 40 years old.")
print("6. Fare values showed a wide variation among passengers.")
print("7. The heatmap shows the relationship between numerical features.")
```
